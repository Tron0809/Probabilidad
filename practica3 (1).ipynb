```python
#Estadistica de los corredores
tiempos=[17.20, 15.10,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]
print(tiempos)
```

```python
mediana = median(tiempos100mts)
print("La media de los tiempos de los corredores es=>",mediana)
```

```python
moda= mode(tiempos)
print("La moda de los tiempos es=>",moda)
```

```python
multimodas = multimode (tiempos)
print("La moda de los tiempos es =>",[multimodas])
```
