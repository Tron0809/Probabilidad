```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Cargar el conjunto de datos
df1 = pd.read_csv("grupo1.csv")
print(df1.head())
```

```python
# Crear un histograma de calificaciones
plt.hist(df1["Calificacion"], bins=10, edgecolor="black")
plt.xlabel("Calificacion")
plt.ylabel("Frecuencia")
plt.title("Histograma de Calificaciones (Grupo 1)")
plt.show()
```

```python
# Crear el modelo de regresión lineal
model = LinearRegression()
# Ajustar el modelo a los datos
model.fit(df1[["HorasEstudio"]], df1["Calificacion"])
# Obtener la ecuación de la línea de regresión
pendiente = model.coef_[0]
intercepto = model.intercept_
print(f"Ecuación de la línea de regresión: Calificación = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
```

```python
#Visualizacion de resultados
plt.scatter(df1["HorasEstudio"], df1["Calificacion"], label="Datos")
plt.plot(df1["HorasEstudio"], color="red", label="Línea de Regresión")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificación")
plt.title("Relación entre Horas de Estudio y Calificación (Grupo 1)")
plt.legend()
plt.show()

```

# Coeficiente de Regresión:
La pendiente en la ecuación de la línea de regresión representa el cambio promedio en la calificación final por cada unidad adicional de tiempo de estudio.
Si la pendiente es positiva, implica que más tiempo de estudio se relaciona con un aumento en la calificación final.
Si la pendiente es negativa, sugiere que más tiempo de estudio se asocia con una disminución en la calificación.


# Coeficiente de Correlación de Pearson:
Mide la fuerza y dirección de la relación lineal entre las variables.
Valores cercanos a 1 indican una fuerte correlación positiva (más estudio, mejores calificaciones).
Valores cercanos a -1 indican una fuerte correlación negativa (más estudio, calificaciones más bajas).
Valores cercanos a 0 sugieren que no hay una relación lineal significativa entre las variables.

# Significancia del Modelo de Regresión Lineal:
Evaluamos la significancia del modelo mediante el valor p asociado al coeficiente de regresión.
Si el valor p es menor que un nivel de significancia (por ejemplo, 0.05), rechazamos la hipótesis nula y afirmamos que el modelo es estadísticamente significativo.
Si el valor p es mayor, no tenemos suficiente evidencia para afirmar que el modelo es significativo.

# Confiabilidad del Modelo de Regresión:
La confiabilidad del modelo depende de varios factores:
Tamaño de la muestra: Un tamaño de muestra más grande generalmente mejora la confiabilidad.
Calidad de los datos: Datos limpios y representativos son esenciales.
Validez de las suposiciones del modelo: Verificar si se cumplen las condiciones de regresión lineal.
El coeficiente de determinación (R²) mide la capacidad del modelo para explicar la variabilidad en las calificaciones. Cuanto más cercano a 1, mejor.
Validación cruzada y métricas de error (como el error cuadrático medio) evalúan la capacidad predictiva del modelo en datos no utilizados durante el entrenamiento.
