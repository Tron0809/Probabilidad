#Ejemplo 3: Histogramas de 2 grupos
1. En ete ejemplo generaremos 100 numeros aleatorios reales que representan la produccion diara de 100 telares (Medidas en metros)
2. Generar otra lista con la produccion de 100 telares para una segunda fabrica
3. Con estas listas de valores se pretende determinar las medidas de tendencia central, es decir, la media, la moda, la mediana y la desviacion estandar
4. Graficar un histograma para cada fabrica



```python
from statistics import*
```

```python
#Libreria para generar numeros aleatorios
import random
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round( random.uniform(50, 100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio2 = round( random.uniform(50, 100),2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
#Libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python
media =mean(fabrica1)
mediana = median(fabrica1)
moda= mode(fabrica1)

print("Valores del a fabrica 1")
print("La media es =",media)
print("La mediana  es",mediana)
print("la moda es ",moda)
```

```python
media =mean(fabrica2)
mediana = median(fabrica2)
moda= mode(fabrica2)

print("Valores del a fabrica 2")
print("La media es =",media)
print("La mediana  es",mediana)
print("la moda es ",moda)
```

```python
#Generar 2 subplot de 1 un reglon por 2 columnas para desplegar los 2 histogtramas,
#Una al lado de la otra
fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2)
```

```python
#Generar 2 subplot de 1 un reglon por 2 columnas para desplegar los 2 histogtramas,
#Una al lado de la otra
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
#Valor minimo,maximo, rango
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
```

```python
plt.hist(fabrica1,bins=20, alpha=0.2, color="b", edgecolor= "black")
plt.axvline(media1, color="b", linestyle="dashed", linewidth=2, label="Media")
plt.axvline(mediana1, color="r", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda1, color="y", linestyle="dashed", linewidth=2, label="Moda")
plt.xlabel=("Valor")
plt.ylabel("Frecuencia")
plt.title("Produccin diara de la fabrica 1")
plt.legend()
plt.show()
```

```python
#Valor minimo,maximo, rango
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

plt.hist(fabrica2,bins=10, alpha=0.2, color="b", edgecolor= "black")
plt.axvline(media2, color="b", linestyle="dashed", linewidth=2, label="Media")
plt.axvline(mediana2, color="r", linestyle="dashed", linewidth=2, label="Mediana")
plt.axvline(moda2, color="y", linestyle="dashed", linewidth=2, label="Moda")
plt.xlabel=("Valor")
plt.ylabel("Frecuencia")
plt.title("Produccin diara de la fabrica 2")
plt.legend()
plt.show()
```

## Tablas de frecuencias

```python
# Generamos una lista vacia que reprenta la edad de un grupo de personas
edades=[]

for i in range(100):
  numero_aleatorio=random.randint(18,70)
  edades.append(numero_aleatorio)
  print(edades)
```

```python
import pandas as pd
```

```python
df_edades=pd.DataFrame(edades, columns=["Edad"])
df_edades
```

```python
#Valcular la tabla de distribucion de frencuencias
tabla_frecuencias= df_edades["Edad"].value_counts().reset_index()
tabla_frecuencias.columns = ["Edad", "Frecuencia"]
tabla_frecuencias=tabla_frecuencias.sort_values(by = "Edad")
tabla_frecuencias
```

```python
tabla_frecuencias["Frec.relativa"]=tabla_frecuencias["Frecuencia"]/100
```

```python
tabla_frecuencias
```

```python
#Calcular la frecuencia porceuntual
tabla_frecuencias["Frecuenias.porcentual"]= tabla_frecuencias["Frecuencia"]/100*100
```

```python
tabla_frecuencias
```

```python
#Calcular la frencuaencia acumulada
tabla_frecuencias["Frec. acumulada"]= tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```

## Regla de Sturges
1-. Es una regla que srive para calcular el numero de clases  o intervalos idneo, en lo que se debe dividir un conjunto de datos
$
c=1+log_2(N)
$

```python
#Libreria de funciones matematicas
import math
```

```python
print (edades)
```

```python
plt.hist(edades,alpha=0.5, rwidth=0.95)
plt.title("Histograma basados en valores estandar de matplotlib" )
plt.ylabel("Frecuencia")
```

```python
df_edades_sturges= pd.DataFrame({"Edad":edades})
df_edades_sturges
```

```python
#Calcular el numero de intervaloo o clases utilizando la regla de sturges
n=len(edades)
k= int(1+math.log2(n))
print("Valor de n=",n,"\n Valor de K o numeros de intervalos o clases=",k)
```

```python
tabla_frec_sturges=pd.cut(df_edades_sturges["Edad"],bins=k).value_counts().reset_index()
tabla_frec_sturges
```

```python
tabla_frec_sturges.columns=["Clase","Frecuencia"]
tabla_frec_sturges
```

```python
#Ordebar la tabla de frecuencias por la clase o intervalos
tabla_frec_sturges = tabla_frec_sturges.sort_values(by ="Clase")
tabla_frec_sturges
```

```python
tabla_frec_sturges["Frec.Relativa"]= tabla_frec_sturges["Frecuencia"]/n
tabla_frec_sturges["Frec.porcentual"]= tabla_frec_sturges["Frecuencia"]/n*100
tabla_frec_sturges["Frec.acumulada"]= tabla_frec_sturges["Frecuencia"].cumsum()
```

```python
tabla_frec_sturges
```

```python
plt.hist(df_edades_sturges["Edad"], bins=k,alpha=0.5,rwidth=.95)
plt.title("HISTOGRAMA DE EDADES BASADO EN LA REGLA DE STURGES")
plt.ylabel("Frecuencia")
```
